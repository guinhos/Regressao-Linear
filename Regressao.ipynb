#### Função para realizar uma análise exploratória de dados (Exploratory Data Analysis - [EDA](https://towardsdatascience.com/exploratory-data-analysis-8fc1cb20fd15)).

```python
def EDA (df):
    eda_df = {}
    eda_df['Amount_NaN'] = df.isnull().sum()
    eda_df['%_NaN'] = df.isnull().mean().round(2)
    eda_df['DType'] = df.dtypes
    eda_df['Amount_Data'] = df.count()
    eda_df['Duplicated'] = df.duplicated().sum()
    
    # Outro ponto para ser verificado, porque para criar a coluna com a quantidade de valores unicos por coluna
    # Não utilizei a função df.unique() 
    colunas = sorted(df.columns.tolist(),
                     key=str.lower, 
                     reverse = False
                    )
    eda_df['Amount_Unique'] = filtered_result = list(map(lambda x: len(df[x].unique().tolist()), colunas))
    
    eda_df['Mean'] = np.round(df.mean(), 2)
    eda_df['Median'] = np.round(df.median(), 2)
    
    eda_df['Max'] = df.max()
    eda_df['Min'] = df.min()
    eda_df['STD'] = np.round(df.std(), 2)
    
    return pd.DataFrame(eda_df)
```

#### Dividir o `dataset` entre um subconjunto de treino e um de teste. Para isso a biblioteca [sklearn.model_selection.train_test_split], que divide arranjos ou matrizes em subconjuntos aleatórios de treino e teste.


#### Aléndisso, isolamo o contunjo alvo (`target`).

```python
from sklearn.model_selection import train_test_split
y = df['price']
del df['price']
```

#### A divisão dos subconjuntos.

```python
x_treino, x_teste, y_treino, y_teste = train_test_split(df.values,
                                                        y,
                                                        test_size = 0.3
                                                       )
```

```python
x_treino.shape
```

#### Saída da divisão entre treino e teste será um arranjo, vamos convertelos em `dataframes` para fazer uma transformação.

```python
x_teste_df = pd.DataFrame(x_teste, 
                          columns = df.columns.tolist()
                         )
x_treino_df = pd.DataFrame(x_treino, 
                           columns = df.columns.tolist()
                          )
```

```python
x_treino_df.head()
#x_teste_df.head
```

```python
frequencia_df = pd.DataFrame()
frequencia_df['neighbourhood'] = x_treino_df['neighbourhood'].value_counts(normalize = True).index.tolist()
frequencia_df['frequencia_neighbourhood'] = x_treino_df['neighbourhood'].value_counts(normalize = True).tolist()
frequencia_df.head()
```

#### A junção dos `dataframes` `x_treino_df` e `frequencia_df` com o método [.merge()](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.merge.html) que mescla `DataFrames` ou objetos nomeados do tipo Series com uma junção de estilo de banco de dados.

```python
x_treino_df = x_treino_df.merge(frequencia_df, 
                                on = 'neighbourhood'
                               )
x_treino_df.head()
```
